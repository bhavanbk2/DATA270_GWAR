```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Identify numeric columns for imputation and scaling
numeric_columns = df_y.select_dtypes(include=[np.number]).columns

# Drop non-numeric columns
df_numeric = df_y[numeric_columns]

# Impute missing values and scale the numeric data
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()

df_numeric_imputed_scaled = pd.DataFrame(scaler.fit_transform(imputer.fit_transform(df_numeric)), columns=numeric_columns)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_numeric_imputed_scaled)

# Create a DataFrame with principal components
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Cluster the data using k-means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(df_numeric_imputed_scaled)
clusters = kmeans.labels_

# Add cluster information to the PCA DataFrame
pca_df['Cluster'] = clusters

# Visualize the score plot with red and blue colors
plt.figure(figsize=(10, 6))

# Assign red (Cluster 0) and blue (Cluster 1) colors
colors = np.where(pca_df['Cluster'] == 0, 'red', 'blue')

plt.scatter(pca_df['PC1'], pca_df['PC2'], c=colors, s=50, alpha=0.5)

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Score Plot with Clustered Categories')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame with both numeric and non-numeric columns
# Select only numeric columns for PCA
numeric_df = df_y.select_dtypes(include=['float64', 'int64'])

# Standardize the numeric data
numeric_df_standardized = StandardScaler().fit_transform(numeric_df)

# Perform PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(numeric_df_standardized)

# Variance explained by each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

# Print or use the resulting DataFrame (pc_df)
print(pc_df)

# Plotting

# 2D Scatter Plot for PC1 and PC2
fig, ax1 = plt.subplots(figsize=(6, 6))
colors = np.where(pca_df['Cluster'] == 0, 'red', 'blue')
ax1.scatter(pc_df['PC1'], pc_df['PC3'], c=colors, s=50, alpha=0.5)
ax1.set_xlabel('Principal Component 1')
ax1.set_ylabel('Principal Component 3')
ax1.set_title('2D Scatter Plot of PCA Results')
plt.show()

# 3D Scatter Plot for PC1, PC2, and PC3
fig = plt.figure(figsize=(8, 8))
ax2 = fig.add_subplot(111, projection='3d')
colors = np.where(pca_df['Cluster'] == 0, 'red', 'blue')
ax2.scatter(pc_df['PC1'], pc_df['PC2'], pc_df['PC3'], c=colors, marker='o')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
ax2.set_zlabel('Principal Component 3')
ax2.set_title('3D Scatter Plot of PCA Results')
plt.show()

# Side panel showing the variance explained by each component
fig, ax3 = plt.subplots(figsize=(3, 6))
bars = ax3.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, align='center')
ax3.set_xticks(range(1, len(explained_variance_ratio) + 1))
ax3.set_ylabel('Variance Explained')
ax3.set_title('Explained Variance by Principal Component')

# Annotate the bars with variance values
for bar, var in zip(bars, explained_variance_ratio):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width() / 2, height, f'{var:.2f}', ha='center', va='bottom')

plt.show()
```
